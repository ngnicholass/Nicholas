```python
# Import library yang diperlukan
import pandas as pd         # Untuk manipulasi data
import numpy as np          # Untuk operasi numerik
import matplotlib.pyplot as plt  # Untuk visualisasi
import seaborn as sns       # Untuk visualisasi yang lebih interaktif dan menarik
```

```python
# 1. Memuat Dataset
# Dataset Titanic tersedia secara built-in di library Seaborn
titanic = sns.load_dataset('titanic')

# Menampilkan 5 baris pertama dataset untuk memahami struktur data
print("5 Baris Pertama Dataset Titanic:")
print(titanic.head())
```

```python
# 2. Informasi Dasar dan Statistik Deskriptif
print("\nInformasi Dataset:")
print(titanic.info())

print("\nStatistik Deskriptif untuk Kolom Numerik:")
print(titanic.describe())
```

```python
# 3. Analisis Missing Values
# Menghitung jumlah missing values per kolom
missing_values = titanic.isnull().sum()
print("\nJumlah Missing Values per Kolom:")
print(missing_values)

# Visualisasi missing values menggunakan heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
plt.title("Heatmap Missing Values pada Dataset Titanic")
plt.show()
```

```python
# 4. Analisis Variabel Kategorik
# Distribusi jumlah penumpang berdasarkan jenis kelamin
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=titanic, palette='pastel')
plt.title("Distribusi Penumpang Berdasarkan Jenis Kelamin")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Jumlah Penumpang")
plt.show()

# Distribusi status kelangsungan hidup (Survived)
plt.figure(figsize=(8, 6))
sns.countplot(x='survived', data=titanic, palette='Set2')
plt.title("Distribusi Status Kelangsungan Hidup Penumpang")
plt.xlabel("Status (0 = Tidak Selamat, 1 = Selamat)")
plt.ylabel("Jumlah Penumpang")
plt.show()
```

```python
# Distribusi jumlah penumpang berdasarkan kelas
plt.figure(figsize=(8, 6))
sns.countplot(x='pclass', data=titanic, palette='cool')
plt.title("Distribusi Penumpang Berdasarkan Kelas")
plt.xlabel("Kelas Penumpang (1 = Atas, 2 = Menengah, 3 = Bawah)")
plt.ylabel("Jumlah Penumpang")
plt.show()

# Distribusi jumlah penumpang berdasarkan pelabuhan embarkasi
plt.figure(figsize=(8, 6))
sns.countplot(x='embarked', data=titanic, palette='muted')
plt.title("Distribusi Penumpang Berdasarkan Pelabuhan Keberangkatan")
plt.xlabel("Pelabuhan (C = Cherbourg, Q = Queenstown, S = Southampton)")
plt.ylabel("Jumlah Penumpang")
plt.show()

# Distribusi kelangsungan hidup berdasarkan kelas
plt.figure(figsize=(8, 6))
sns.countplot(x='pclass', hue='survived', data=titanic, palette='Set1')
plt.title("Distribusi Kelangsungan Hidup Berdasarkan Kelas")
plt.xlabel("Kelas Penumpang")
plt.ylabel("Jumlah Penumpang")
plt.legend(title="Survived", labels=["Tidak Selamat", "Selamat"])
plt.show()

# Distribusi kelangsungan hidup berdasarkan pelabuhan embarkasi
plt.figure(figsize=(8, 6))
sns.countplot(x='embarked', hue='survived', data=titanic, palette='Set2')
plt.title("Distribusi Kelangsungan Hidup Berdasarkan Pelabuhan Keberangkatan")
plt.xlabel("Pelabuhan")
plt.ylabel("Jumlah Penumpang")
plt.legend(title="Survived", labels=["Tidak Selamat", "Selamat"])
plt.show()
```

```python
# 5. Analisis Variabel Numerik
# Histogram Distribusi Usia (Age)
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribusi Umur Penumpang")
plt.xlabel("Umur")
plt.ylabel("Frekuensi")
plt.show()

# Boxplot Umur berdasarkan Kelas Penumpang (pclass)
plt.figure(figsize=(8, 6))
sns.boxplot(x='pclass', y='age', data=titanic, palette='Set3')
plt.title("Distribusi Umur Berdasarkan Kelas Penumpang")
plt.xlabel("Kelas Penumpang")
plt.ylabel("Umur")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram distribusi umur
sns.histplot(ax=axes[0,0], data=titanic, x='age', bins=30, kde=True, color='skyblue')
axes[0,0].set_title("Distribusi Umur Penumpang")
axes[0,0].set_xlabel("Umur")
axes[0,0].set_ylabel("Frekuensi")

# Histogram distribusi tarif
sns.histplot(ax=axes[0,1], data=titanic, x='fare', bins=40, kde=True, color='orange')
axes[0,1].set_title("Distribusi Tarif Tiket Penumpang")
axes[0,1].set_xlabel("Tarif (Fare)")
axes[0,1].set_ylabel("Frekuensi")

# Boxplot umur berdasarkan kelas
sns.boxplot(ax=axes[1,0], x='pclass', y='age', data=titanic, palette='Set3')
axes[1,0].set_title("Distribusi Umur Berdasarkan Kelas Penumpang")
axes[1,0].set_xlabel("Kelas Penumpang")
axes[1,0].set_ylabel("Umur")

# Boxplot tarif berdasarkan kelas
sns.boxplot(ax=axes[1,1], x='pclass', y='fare', data=titanic, palette='coolwarm')
axes[1,1].set_title("Distribusi Tarif Berdasarkan Kelas Penumpang")
axes[1,1].set_xlabel("Kelas Penumpang")
axes[1,1].set_ylabel("Tarif (Fare)")

plt.tight_layout()
plt.show()
```

```python
# 6. Analisis Korelasi antar Variabel Numerik
# Hanya memilih kolom numerik agar tidak terjadi error konversi
numeric_data = titanic.select_dtypes(include=[np.number])
correlation_matrix = numeric_data.corr()
print("\nMatriks Korelasi:")
print(correlation_matrix)

# Visualisasi matriks korelasi menggunakan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap Matriks Korelasi pada Dataset Titanic")
plt.show()
```

```python
# 7. Visualisasi Hubungan Antar Variabel
# Scatter Plot: Hubungan antara Umur (age) dan Tarif (fare)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', data=titanic, hue='survived', palette='Set1', alpha=0.7)
plt.title("Scatter Plot: Umur vs. Tarif dengan Status Kelangsungan Hidup")
plt.xlabel("Umur")
plt.ylabel("Tarif")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
# Scatter Plot: Umur vs Tarif dengan kelas penumpang
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', data=titanic, hue='pclass', palette='viridis', alpha=0.7)
plt.title("Scatter Plot: Umur vs. Tarif dengan Kelas Penumpang")
plt.xlabel("Umur")
plt.ylabel("Tarif")
plt.legend(title="Kelas Penumpang")
plt.show()

# Boxplot: Umur berdasarkan jenis kelamin dan status kelangsungan hidup
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic, palette='Set2')
plt.title("Boxplot: Umur Berdasarkan Jenis Kelamin dan Status Kelangsungan Hidup")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Umur")
plt.legend(title="Survived", labels=["Tidak Selamat", "Selamat"])
plt.show()

# Scatter Plot: Jumlah keluarga (sibsp + parch) vs Tarif
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1  # +1 untuk diri sendiri
plt.figure(figsize=(8, 6))
sns.scatterplot(x='family_size', y='fare', data=titanic, hue='survived', palette='coolwarm', alpha=0.7)
plt.title("Scatter Plot: Ukuran Keluarga vs Tarif dengan Status Kelangsungan Hidup")
plt.xlabel("Ukuran Keluarga (sibsp + parch + 1)")
plt.ylabel("Tarif")
plt.legend(title="Survived", labels=["Tidak Selamat", "Selamat"])
plt.show()

# Violin Plot: Distribusi Tarif berdasarkan Kelas dan Jenis Kelamin
plt.figure(figsize=(8, 6))
sns.violinplot(x='pclass', y='fare', hue='sex', data=titanic, split=True, palette='muted')
plt.title("Distribusi Tarif Berdasarkan Kelas dan Jenis Kelamin")
plt.xlabel("Kelas Penumpang")
plt.ylabel("Tarif")
plt.show()
```

```python
# 8. Analisis Lanjutan: Pairplot
# Pairplot untuk melihat hubungan antar variabel numerik dan kategorik secara menyeluruh
sns.pairplot(titanic, hue='survived', palette='Set2', vars=['age', 'fare', 'pclass'])
plt.suptitle("Pairplot Variabel pada Dataset Titanic", y=1.02)
plt.show()

# 9. Kesimpulan Sementara dari EDA:
# - Dataset Titanic memiliki beberapa missing values, terutama pada kolom 'age', 'deck', dan 'embark_town'.
# - Distribusi penumpang tampak berbeda berdasarkan jenis kelamin dan status kelangsungan hidup.
# - Variabel 'fare' dan 'age' memiliki sebaran yang cukup bervariasi, dengan outlier yang terlihat di boxplot.
# - Matriks korelasi menunjukkan adanya korelasi antara variabel numerik tertentu, misalnya antara 'fare' dan 'pclass'.
# - Visualisasi hubungan (scatter plot dan pairplot) memberikan wawasan tentang interaksi antar variabel dan perbedaan antara penumpang yang selamat dan tidak.
```
